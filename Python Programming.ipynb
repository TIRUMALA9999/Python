#### Python program to find whether a given number is prime or not

```python
num = int(input('Enter a number: '))
if num > 1:
    for i in range(2, int(num/2)+1):
        # If num is divisible by any number between
        # 2 and n / 2, it is not prime
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
def fibonacci(n):
    a = 0
    b = 1
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        for i in range(2, n):
            c = a + b
            a = b
            b = c
        return b
print(fibonacci(9))
```

#### Python Program for How to check if a given number is Fibonacci number?

```python
## A number is Fibonacci if and only if one or both of (5*n2 + 4) or (5*n2 – 4) is a perfect square
```

```python
# python program to check if x is a perfect square
import math

def isPerfectSquare(x):
    s = int(math.sqrt(x))
    return s*s == x
def isFibonacci(n):
    return isPerfectSquare(5*n*n + 4) or isPerfectSquare(5*n*n - 4)


for i in range(1, 11):
    if (isFibonacci(i) == True):
        print(i, "is a Fibonacci Number")
    else:
        print(i, "is a not Fibonacci Number ")
```

#### Python Program for n\’th multiple of a number in Fibonacci Series

```python
def nth_fib_multiple(n, k):
    fibonacci = [0, 1]
    counter = 0
    for i in range(2, 10000):
        current = fibonacci[i-1] + fibonacci[i-2]
        if current % k == 0:
            counter += 1
            if counter == n:
                return i
        fibonacci.append(current)
         
# Example usage
print(nth_fib_multiple(5, 4)) # Output: 9
```

```python
#Using divide and conquer method
def sum_of_array(arr, low, high):
    if low == high:
        return arr[low]
    mid = (low + high) // 2
    left_sum = sum_of_array(arr, low, mid)
    right_sum = sum_of_array(arr, mid+1, high)
    return left_sum + right_sum
#Examples
arr = [1, 2, 3]
print(sum_of_array(arr, 0, len(arr)-1)) # Output: 6
 
arr = [15, 12, 13, 10]
print(sum_of_array(arr, 0, len(arr)-1)) # Output: 50
```

```python
def largest(arr):
    max2=arr[0]
    for i in range(len(arr)):
        if arr[i]>max2:
            max2=arr[i]
    return max2
```

```python
largest([10,324,45,90,9808])
```
