## self --> Convention
### By using the “self”  we can access the attributes and methods of the class in Python.

```python
class Student:
    def __init__(self,student_name,student_age):
        self.name=student_name   
        self.age=student_age
        print("I am one of the magic method created and called for student class")
    def first_function(self):
        print("I am the first function which should be called")
    def study(self):
        print("I am studying now")
    def play(self):
        print("I am playing now "+str(self.name))
```

### Create a python class named rectangle to represent rectangles . The class should have the following features:
#### A constructor method (__init__) that initialises the rectangles width and height.
#### Two Methods:
#### calculate_area() method that calculates and returns the area of the rectangle.
#### calculate_perimeter() method that calculates and returns the perimeter of the rectangle.
#### Create two objects of the rectangle class and demonstrate the use of these methods.
#### write the code to implement the rectangle class and demonstrate its functionality with given instructions

```python
class Rectangle:
    def __init__(self,width,height):
        self.width=width
        self.height=height
    def calculate_area(self):
        area=0
        area=self.width*self.height
        return print("area of the rectangle is",area)
    def calculate_perimeter(self):
        perimeter=0
        perimeter= 2*(self.width+self.height)
        return print("perimeter of the rectangle is: ",perimeter)
        
```

```python
r1=Rectangle(2.5,3)
```

```python
r1.calculate_area()
```

```python
r2=Rectangle(8,3)
```

```python
r2.calculate_perimeter()
```

### Create a python class named Fraction to represent fractions. The class should have the following features:
#### 1.  A constructor method (init) that initializes the fractions numerator and denominator.
#### 2.  Methods to perform the following operations :
####  Addition(+),  Subtraction(-),  Multiplication(*),  Division(/),  Comparision (==,!=,<,<=,>,>=)
#### 3. A Method to simplify the fraction 

```python
class Fraction:
    def __init__(self,num,den):
        self.numerator=num
        self.denominator=den
        print("A fraction is created")
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"
    def __add__(self,other):
        print(self)
        print(other)
        new_num=self.numerator * other.denominator + other.numerator * self.denominator
        new_den=self.denominator*other.denominator
        self=Fraction(new_num,new_den)
        return self
        
```

```python
f1=Fraction(1,5)
```

```python
f2=Fraction(3,5)
```

```python
print(f1)
```

```python
print(f2)
```
